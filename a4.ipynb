```python

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
dfp = df.toPandas()
f,ax=plt.subplots(1,2,figsize=(18,8))
dfp['crashSeverity'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('crashSeverity Distribution')
ax[0].set_ylabel('Count')
sns.countplot('crashSeverity',data=dfp,ax=ax[1],order=dfp['crashSeverity'].value_counts().index)
ax[1].set_title('Count of crashSeverity')
plt.show()
```
